# RandomForest Implementation | Classification

```python
#Reading all the packages 
import pandas as pd
import numpy as np
import tensorflow_data_validation as tfdv
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier as rf
from sklearn import datasets
```

## Reading the data and visualizing

```python
##Reading a dataset of patient with heart disease to build a predictor which can predict heart disease in people

df = pd.read_csv('D:\Github_projects\Classifier- RandomForest/HeartDisease.csv')
print(df.dtypes)
df.head()
```

## EDA using TensorFlow Data Validation

#### Before starting the EDA, let's split the data in training and testing datasets

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, random_state=1)
```

```python
#creating a complete train dataset for EDA
train_df = pd.concat([X_train, y_train], axis=1)
train_df
```

#### Beginning EDA in TFDV

```python
train_stats = tfdv.generate_statistics_from_dataframe(train_df)
tfdv.visualize_statistics(train_stats)
```

```python
schema = tfdv.infer_schema(statistics=train_stats)
tfdv.display_schema(schema=schema)
```
